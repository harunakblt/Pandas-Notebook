## <font color='blue'> <b>transform()</b><font color='black'>
    
transform(), bir pandas DataFrame ya da Serisi'ndeki verileri deitirmek veya dntrmek iin kullanlan bir yntemdir. 

Belirtilen ilevi uygular ve sonucu yeni bir Seri veya DataFrame olarak dndrr.

zellikle gruplama ilemleri ile birlikte kullanldnda olduka yararldr.

[SOUREC01](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transform.html) & [SOURCE02](https://www.analyticsvidhya.com/blog/2020/03/understanding-transform-function-python/)
    




```python
data = {'Company':['Amazon', 'Amazon', 'OpenAI', 'OpenAI', 'Amazon', 'OpenAI', 'Amazon', 'OpenAI'],
        'Department':['IT', 'HR', 'IT', 'HR', 'IT', 'HR', 'HR', 'IT'],
        'Person':['Ahmet', 'Hatice', 'Fatma', 'Mehmet', 'Zeynep', 'Ali', 'Yusuf', 'Elif'],
        'Age':[32, 26, 35, 40, 25, 38, 30, 28],
       
        'Salary':[180, 160, 150, 180, 240, 200, 220, 160]}
```

```python
df = pd.DataFrame(data = data)
df
```

```python
df.groupby("Department")[["Salary"]].mean()
```

```python
df.groupby("Department")["Salary"].transform("mean")
```

```python
df["departman_ortalamas"] = df.groupby("Department")["Salary"].transform("mean")
```

```python
df
```

```python
df.Salary - df.departman_ortalamas
```

```python
df["ortalamaya_gre_fark"] = df.Salary - df.departman_ortalamas
```

```python
df
```

```python
df.Age.transform(lambda x: x + 5)
```

```python
df.Age = df.Age.transform(lambda x: x + 5)
```

```python
df
```

**None**:
   - Python'da "hibir ey" anlamna gelir.
   - Python'da bir nesnenin deeri olmadn gstermek iin kullanlr.
   - Tipi NoneType'dr ve sadece tek bir deeri vardr: None.
   - rnein, bir fonksiyonun herhangi bir deer dndrmemesi durumunda, Python otomatik olarak None dndrr.
   
**nan ve NaN**:
   - "Not a Number" ifadesinin ksaltmasdr.
   - float tipinde bir deeri temsil eder.
   - Saysal olmayan deerleri gstermek iin kullanlr. 
   - Pandas ve Numpy gibi ktphanelerde, kayp veriyi (missing data) temsil etmek iin de kullanlr.
   - nan deerleri eitlikle karlatrldnda her zaman False dndrr. Yani, nan == nan ifadesi False sonucunu verir. 
   - nan ve NaN arasnda teknik bir fark yoktur. Farkl kaynaklarda veya ktphanelerde farkl yazm biimleri tercih edilir, ancak ikisi de ayn anlama gelir.

**zetle**:
- None Python'n zgn bir deeridir ve "deer yok" anlamna gelir.
- nan veya NaN saysal olmayan deerleri temsil eder ve genellikle kayp veri veya matematiksel olarak tanmlanmayan ilemler iin kullanlr. 

Bu iki deer farkl tiplere aittir ve farkl durumlar iin kullanlr. 

Uygulamada, veri analizi yaparken Pandas'ta ska NaN deerleriyle karlarsnz. Ancak genel Python programlamasnda None daha yaygndr.
